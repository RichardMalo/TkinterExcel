```python
import tkinter as tk
from tkinter import messagebox
import csv
from datetime import datetime

class Application(tk.Frame):
    def __init__(self, master=None):
        super().__init__(master)
        self.master = master
        self.pack()
        self.create_widgets()

    def create_widgets(self):
        self.first_name_label = tk.Label(self, text="First Name:")
        self.first_name_label.pack()

        self.first_name_entry = tk.Entry(self, width=40)  # Increase width to 40
        self.first_name_entry.pack()

        self.last_name_label = tk.Label(self, text="Last Name:")
        self.last_name_label.pack()

        self.last_name_entry = tk.Entry(self, width=40)  # Increase width to 40
        self.last_name_entry.pack()

        self.dob_label = tk.Label(self, text="Date of Birth (yyyy-mm-dd):")
        self.dob_label.pack()

        self.dob_entry = tk.Entry(self, width=40)  # Increase width to 40
        self.dob_entry.pack()

        self.marriage_status_label = tk.Label(self, text="Marriage Status:")
        self.marriage_status_label.pack()

        self.marriage_status_entry = tk.Entry(self, width=40)  # Increase width to 40
        self.marriage_status_entry.pack()

        self.submit_button = tk.Button(self, text="SUBMIT", command=self.write_to_csv)
        self.submit_button.pack()

    def write_to_csv(self):
        fieldnames = ['First Name', 'Last Name', 'Date of Birth', 'Marriage Status']

        # verify if date format is valid
        try:
            datetime.strptime(self.dob_entry.get(), '%Y-%m-%d')
        except ValueError:
            messagebox.showerror("Invalid input", "Date of Birth must be in format yyyy-mm-dd")
            return

        with open('output.csv', 'a', newline='') as file:
            writer = csv.DictWriter(file, fieldnames=fieldnames)

            # writer.writeheader()  # uncomment this if you want headers in your CSV

            writer.writerow({
                'First Name': self.first_name_entry.get(),
                'Last Name': self.last_name_entry.get(),
                'Date of Birth': self.dob_entry.get(),
                'Marriage Status': self.marriage_status_entry.get()
            })

        messagebox.showinfo("Success", "Data successfully saved to output.csv")

        # Clear the inputs
        self.first_name_entry.delete(0, 'end')
        self.last_name_entry.delete(0, 'end')
        self.dob_entry.delete(0, 'end')
        self.marriage_status_entry.delete(0, 'end')

root = tk.Tk()
app = Application(master=root)
app.mainloop()
```
